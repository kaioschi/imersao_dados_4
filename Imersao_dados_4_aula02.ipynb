```python
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]  # criando um novo dataframe, usando apenas o slice com Tipo_anuncio vazio, para remover as linhas do df que contém informações como /Mês e /Ano, já que queremos apeans verificar imóveis para venda, e não aluguel
dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)  # o df usava . como separador de decimais, mas o padrão do python não é esse, então usamos str.replace() para trocar por vazio, além de usar astype() para converter para float (que já vai colocar o ponto no lugar certo)
```

```python
dados_vendas["Valor_anuncio_float"] = dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)  # passando o acerto da casa decimal e conversão de tipo para uma nova coluna do df dados_vendas
```

# Desafio 3 

Preço do metro quadrado por bairro e plotar em um gráfico ideal

```python
dados_vendas["Preço m²"] = dados_vendas["Valor"] / dados_vendas["Metragem"]
preco_m2 = dados_vendas.groupby("Bairro")["Preço m²"].mean().to_frame().reset_index().sort_values(by = "Preço m²", ascending = False).head(20)
preco_m2
```

```python
sns.set()
plt.figure(figsize = (20, 8))
ax = sns.barplot(data = preco_m2, x = "Bairro", y = "Preço m²")
ax.bar_label(ax.containers[0], label_type='edge') # colocando labels com os valores, o ax.containers[0] é o default do método, não pesquisei muito sobre ele
# ax.get_yaxis().set_visible(False) usar esse remove todo o eixo y, até o label
ax.get_yaxis().set_ticks([]) # deixando os ticks do eixo y vaizos, ou seja, sem os números
plt.xticks(rotation = 70) # inclinando os ticks do eixo x em 70 graus
plt.tick_params(colors = "gray") # alterando parâmetros dos ticks
plt.xlabel('Bairros', fontsize = 10)
plt.ylabel('Preço por m²', fontsize = 10)
plt.title("Preço por m² dos bairros de São Paulo", fontsize = 16)
```

O preço médio tem uma discrepância pois a base de dados ainda possui os valores de imóveis que tinham alugueis por mês e ano, apenas as strings que denotavam esse tipo de aluguel foram removidas, os valores permaneceram cheios, e por isso, distorcem o preço médio por m².
