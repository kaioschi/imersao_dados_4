# Desafio 1

Realizar a média da metragem para cada um dos bairros (Paulo)

```python
# utilizando o método groupby() agrupando Bairro e selecionando Metragem com o método mean()

dados.groupby("Bairro")["Metragem"].mean()
```

# Desafio 2

2 formas de selecionar os dados por bairro (consultar os métodos na documentação do pandas) (Thiago)


```python
# utilizando loc para selecionar bairros individualmente

dados.loc[dados["Bairro"] == "Mooca"].sample(5)
```

```python
# selecionando as primeiras 10 linhas onde a coluna Bairro é igual a Mooca e coluna Quartos é igual ou menor a 3

dados.loc[(dados["Bairro"] == "Liberdade") & (dados["Quartos"] <= 3)].head(5)
```

```python
# utilizando o iloc para selecionar as primeiras 10 linhas e as segunda, terceira e quarta colunas do dataframe, ordenando por ordem alfabética dos bairros

dados.iloc[:10, 1:4].sort_values(by=["Bairro"], ascending=True)
```

# Desafio 3 

Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão (Thiago)

```python
# criando um novo dataframe com as colunas todas as linhas de dados, mas apenas as colunas de Bairro e Quartos

numero_quartos = pd.DataFrame(dados.iloc[:, [1, 4]])
numero_quartos.dropna(inplace=True)
numero_quartos
```

```python
# contando quantos imóveis tem certas quantidades de quartos, mas ele usa Quartos como index do dataframe, WHY?

qtd_imoveis_quartos = pd.DataFrame(data=numero_quartos.groupby("Quartos")["Bairro"].count()).reset_index()
qtd_imoveis_quartos.columns = ["Quartos", "Número de Imóveis"]
qtd_imoveis_quartos
```

```python
qtd_imoveis_quartos.head(5).plot.line()
```

```python
# criando uma series filtrando apenas o nome do bairro desejado para o gráfico

bairro_escolhido = numero_quartos["Bairro"] == "Mooca"
bairro_escolhido
```

```python
# contando a quantidade de de imóveis com base no número de quartos que possuem

grafico = pd.DataFrame(numero_quartos[bairro_escolhido].value_counts(sort=False)).reset_index()
grafico.columns = ("Bairro", "Nº de Quartos", "Qtde. Imóveis")
grafico
```

```python
grafico.plot.line(y="Nº de Quartos", x="Qtde. Imóveis")
```

# Desafio 4 

Pegar outros estatísticas dos dados, média, mediana, mim, max (Vivi)

```python
# trazendo as estatística da base de dados, incluíndo quartis

estatisticas = dados.groupby("Bairro")["Metragem"].describe()
estatisticas
```

```python
# mostrando os imóveis com maior metragem em cada bairro em ordem alfabética, porém temos caracteres especiais (letras com acentos), que não são ordenados corretamente (ainda)
# podemos trocar pelos métodos min() para menor metragem, mean() para média e std() para desvio padrão

dados_max = dados.groupby("Bairro")["Metragem"].max()
dados_max.sort_values(ascending=False)
```

# Desafio 5 

Descobrir quais são os bairros que não tem nome de Rua. (Vivi)

```python
# usando o método isna() para encontrar as ruas de dados que são NaN e incluindo essa seleção em uma variável

ruas_sem_nome = dados["Rua"].isna()
ruas_sem_nome
```

```python
# criando outra variável com base na seleção booleana de ruas_sem_nome, e usando o método unique() para remover os nomes de bairros que aparecem multiplas vezes

bairros_sem_rua = dados[ruas_sem_nome]["Bairro"].unique()
bairros_sem_rua
```

```python
# criando um novo dataframe com nomes únicos de cada bairro sem nomes de ruas

bairros_unicos = pd.DataFrame(bairros_sem_rua)
bairros_unicos
```
